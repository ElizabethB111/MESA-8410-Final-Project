```python
# Merge the DataFrames into a Single DataFrame For the Analysis;

# First - Perform OUTER merge of Police Base and Full

# Merge together into df3, using id
police_outer = pd.merge(police_base_data, police_full_data, on = 'uid', how = "outer")


# Create ps_code by coelescing ps_code_x and ps_code_y to enable join to personnel data;

police_outer['ps_code'] = police_outer.ps_code_x.combine_first(police_outer.ps_code_y)


# Look at the dimensions of the data after merging
print("police_outer dimensions (rows, cols): ", police_outer.shape)
```

```python
# Append Police Personnel Data Onto Each Record of Police Outer;


# Merge together into df3, using id
police_outer_person = pd.merge(police_outer, police_person_data, on = 'ps_code', how = "left")

# Look at the dimensions of the data after merging
print("police_outer_person dimensions (rows, cols): ", police_outer_person.shape)

```
