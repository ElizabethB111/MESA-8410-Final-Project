# Questions 1 and 2: Are Women's Help Desks more effective when run by women or when they are regular help desks (70% of which are run by men)? Does the presence of a Women's Help Desk in a police station increase protection of women?
# These two questions were combined into the tables below.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {'Treatment': ['Control', 'Women-Run WHD', 'Regular WHD'],
         'Mean Endline FIR Overall Count': [32.28, 32.36, 33.58],
         'Mean Endline FIR CAW Count': [3.84, 4.1, 4.97],
         'Mean Endline FIR By Women': [2.6, 2.72, 3.13],
         'Mean Endline Arrest Count': [3.54, 3.6, 4.12],
         'Mean Endline DIR Count': [.03, .91, .89]}
df = pd.DataFrame(data)
```

The data above came from the Admin Wide Station Level Means data set created in previous checkpoints, moved into excel to create a pivot table for averages, then brought back into this notebook.

```python

data = {'Treatment': ['Control', 'Women-Run WHD', 'Regular WHD'],
         'Mean Endline FIR Overall Count': [32.28, 32.36, 33.58],
         'Mean Endline FIR CAW Count': [3.84, 4.1, 4.97],
         'Mean Endline FIR By Women': [2.6, 2.72, 3.13],
         'Mean Endline Arrest Count': [3.54, 3.6, 4.12],
         'Mean Endline DIR Count': [.03, .91, .89]}
df = pd.DataFrame(data)

df.set_index('Treatment', inplace=True)
df_transposed = df.T

ax = df_transposed.plot(kind='bar', figsize=(10, 6))

plt.title('Comparison of Report Numbers Between Treatment Effects')
plt.ylabel('Mean Values')
plt.xlabel('Type of Report')
plt.xticks(rotation=45)
plt.legend(title='Metrics')
plt.tight_layout()
plt.show()
```

```python
data = {'Treatment': ['Control', 'Women-Run WHD', 'Regular WHD'],
         'Mean Endline FIR Overall Count': [32.28, 32.36, 33.58],
         'Mean Endline FIR CAW Count': [3.84, 4.1, 4.97],
         'Mean Endline FIR By Women': [2.6, 2.72, 3.13],
         'Mean Endline Arrest Count': [3.54, 3.6, 4.12],
         'Mean Endline DIR Count': [.03, .91, .89]}
df = pd.DataFrame(data)

fig, axs = plt.subplots(3, 2, figsize=(15, 15))

sns.barplot(x='Treatment', y='Mean Endline FIR Overall Count', data=df, ax=axs[0, 0])
axs[0, 0].set_title('Mean Endline FIR Overall Count')

sns.barplot(x='Treatment', y='Mean Endline FIR CAW Count', data=df, ax=axs[0, 1])
axs[0, 1].set_title('Mean Endline FIR CAW Count')

sns.barplot(x='Treatment', y='Mean Endline FIR By Women', data=df, ax=axs[1, 0])
axs[1, 0].set_title('Mean Endline FIR By Women')

sns.barplot(x='Treatment', y='Mean Endline Arrest Count', data=df, ax=axs[1, 1])
axs[1, 1].set_title('Mean Endline Arrest Count')

sns.barplot(x='Treatment', y='Mean Endline DIR Count', data=df, ax=axs[2, 0])
axs[2, 0].set_title('Mean Endline DIR Count')

fig.delaxes(axs[2, 1])

plt.tight_layout()
plt.show()
```

Below is another alternative way to graph. This graph below makes it more difficult to tell the differences between each group within each report type because the y-axis numbers are spread further. I would be less likely to use this graph below.

```python
data = {'Treatment': ['Control', 'Women-Run WHD', 'Regular WHD'],
            'Mean Endline FIR Overall Count': [32.28, 32.36, 33.58],
            'Mean Endline FIR CAW Count': [3.84, 4.1, 4.97],
            'Mean Endline FIR By Women': [2.6, 2.72, 3.13],
            'Mean Endline Arrest Count': [3.54, 3.6, 4.12],
            'Mean Endline DIR Count': [.03, .91, .89]}
df = pd.DataFrame(data)

df_melted = pd.melt(df, id_vars=['Treatment'], var_name='Metric', value_name='Value')

plt.figure(figsize=(12, 6))
sns.barplot(x='Treatment', y='Value', hue='Metric', data=df_melted)
plt.title('Comparison of Report Types Across Treatment Groups')
plt.ylabel('Mean Values')
plt.xticks(rotation=45)
plt.legend(title='Metrics', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

The heatmap below is fun to examine, and clearer than the graph above.

```python
data = {'Treatment': ['Control', 'Women-Run WHD', 'Regular WHD'],
            'Mean Endline FIR Overall Count': [32.28, 32.36, 33.58],
            'Mean Endline FIR CAW Count': [3.84, 4.1, 4.97],
            'Mean Endline FIR By Women': [2.6, 2.72, 3.13],
            'Mean Endline Arrest Count': [3.54, 3.6, 4.12],
            'Mean Endline DIR Count': [.03, .91, .89]}
df = pd.DataFrame(data)

df.set_index('Treatment', inplace=True)

plt.figure(figsize=(8, 6))
sns.heatmap(df, annot=True, cmap='viridis', fmt=".2f")
plt.title('Heatmap of Treatment Effects')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Treatment': ['Control', 'Women-Run WHD', 'Regular WHD'],
        'Mean Endline FIR CAW Count': [3.84, 4.1, 4.97],
        'Mean Endline FIR By Women': [2.6, 2.72, 3.13],
        'Mean Endline Arrest Count': [3.54, 3.6, 4.12],
        'Mean Endline DIR Count': [.03, .91, .89]}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
for metric in df.columns[1:]:
    plt.plot(df['Treatment'], df[metric], marker='o', label=metric)

plt.title('Comparison of Reports Across Treatment Groups')
plt.xlabel('Treatment Group')
plt.ylabel('Mean Value')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.tight_layout()
plt.show()
```
